```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page_url = 'https://www.otodom.pl/pl/wyniki/sprzedaz/mieszkanie/mazowieckie/warszawa/warszawa/warszawa?limit=36&ownerTypeSingleSelect=ALL&priceMax=300000&by=DEFAULT&direction=DESC&viewType=listing'
page = requests.get(page_url)
print(page)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
list_of_url = []
for a in soup.find_all('a', {"class": "css-lsw81o e1dfeild2"}, href=True):
    one_url = a['href']
    list_of_url.append(one_url)
```

```python
list_of_titles = []
adds_titles = soup.find_all("span", {"class": "css-1xq1a7w e1n06ry52"})
for i in adds_titles: 
    one_title = i['title']
    list_of_titles.append(one_title)
```

```python
list_of_subtitles = []
adds_subtitles = soup.find_all("p", {"class": "css-19dkezj e1n06ry53"})
for i in adds_subtitles: 
    one_subtitle = i['title']
    list_of_subtitles.append(one_subtitle)
```

```python
# szczegoly ogloszen - cena, cena za metr , liczba pokoi, liczba metrow

list_of_details_all_adds = []

all_details = soup.find_all("div", {"class": "e1jyrtvq0"})

for one_add_detail in all_details:
    list_of_one_add_details = []
    
    for one_info in one_add_detail.select("span",  attr="text"):
        one_detail = one_info.text
        list_of_one_add_details.append(one_detail)
        
    list_of_details_all_adds.append(list_of_one_add_details)

one_page_details_df = pd.DataFrame(list_of_details_all_adds, columns = ['Price', 'm2price', 'rooms', 'm2']) 
```

```python
one_page_details_df['url'] = list_of_url
one_page_details_df['titles'] = list_of_titles
one_page_details_df['subtitles'] = list_of_subtitles
```
